## Ejercicio 5- Calcular logaritmos apartir de raices.

```python

import math
```

```python
valoReal=round(math.log10(2),7)
# constante α = 1 / ln(10)
alpha = 1 / math.log(10)
n=32
def calculo_logaritmos(): #Primera version (solo imprime la aproximacion)
    valores=[]
    for i in range (1,n+1):
        r=(2**(1/(2**i)))-1
        log_approx=alpha*r
        valor=round(log_approx,6)
        valores.append(valor)
    log_briggs=sum(valores)
    return log_briggs
 
```

```python

calculado=calculo_logaritmos()
print("El valor aproximado de log10(2) es:",calculado)
print("El valor real de log10(2) es:",valoReal)
```

```python
def log_briggs(x, k=5, precision=6):
    alpha = 1 / math.log(10)
    xn = float(x)
    # Aplicar k raíces cuadradas consecutivas
    for _ in range(k):
        xn = round(math.sqrt(xn), precision)
    r = xn - 1
    approx_log = alpha * r
    approx_log *= (2 ** k)

    return approx_log, math.log10(x)

numeros = [2, 3, 4, 5, 6, 7, 8, 9, 10 ]

print(f"{'Número':<8} {'Log Método Briggs':<22} {'Log real':<18} {'Error absoluto'}")
print("-"*75)

for n in numeros:
    lb, real = log_briggs(n)
    print(f"{n:<8} {lb:<22.6f} {real:<18.6f} {abs(lb-real):.6f}")
```

```python

from matplotlib import pyplot as plt


def graficar_log6(k_max=20, precision=10):
    """
    Grafica la aproximación de log10(6) usando el método de Briggs para k=0..k_max
    Muestra la aproximación vs k y el error absoluto vs k.
    """
    valores_k = []
    aproximaciones = []
    errores = []

    verdadero = math.log10(6)
    for k in range(0, k_max+1):
        approx, _ = log_briggs(6, k=k, precision=precision)
        valores_k.append(k)
        aproximaciones.append(approx)
        errores.append(abs(approx - verdadero))

    # Gráfica de aproximación vs k con línea del valor real
    plt.figure(figsize=(8,4))
    plt.plot(valores_k, aproximaciones, marker='o', label='Aproximación Briggs')
    plt.hlines(verdadero, valores_k[0], valores_k[-1], colors='r', linestyles='--', label='log10(6) real')
    plt.xlabel('Número de raíces k')
    plt.ylabel('log10(6)')
    plt.title('Aproximación de log10(6) vs número de raíces (k)')
    plt.grid(True)
    plt.legend()
    plt.show()

    # Gráfica del error absoluto vs k
    plt.figure(figsize=(8,4))
    plt.plot(valores_k, errores, marker='s', color='magenta', label='Error absoluto')
    plt.yscale('log')  # opcional: escala log para ver mejor la convergencia
    plt.xlabel('Número de raíces k')
    plt.ylabel('Error absoluto')
    plt.title('Error absoluto en la aproximación de log10(6)')
    plt.grid(True, which='both', ls='--')
    plt.legend()
    plt.show()

```

```python
graficar_log6(k_max=20, precision=10)
```
